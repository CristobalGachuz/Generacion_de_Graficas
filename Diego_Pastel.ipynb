```python
import matplotlib.pyplot as plt

# Datos de ejemplo
categorias = ['Manzanas', 'Bananas', 'Cerezas', 'Uvas']
valores = [25, 30, 20, 25]

# Crear el gráfico
plt.figure(figsize=(6, 6))
plt.pie(valores, labels=categorias, autopct='%1.1f%%', startangle=90)
plt.title('Distribución de Frutas')
plt.axis('equal')  # Para que sea un círculo

# Mostrar gráfico
plt.show()
```
