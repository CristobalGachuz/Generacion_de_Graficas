### Volver al índice
1. [Índice](https://github.com/CristobalGachuz/Generacion_de_Graficas/blob/main/indice.ipynb)

```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.DataFrame({
    'LimInf': [0, 10, 20, 30, 40],
    'LimSup': [10, 20, 30, 40, 50],
    'Frecuencia': [5, 10, 15, 20, 10]
})

data['FreeAcum'] = data['Frecuencia'].cumsum()

plt.figure(figsize=(10, 6))
plt.plot(data['LimSup'], data['FreeAcum'], 'o-', label='ojiva')
plt.plot([data['LimInf'].iloc[0]] + data['LimSup'].tolist(),
         [0] + data['FreeAcum'].tolist(), 'o-')

plt.title('Ojiva de Frecuencias Acumuladas')
plt.xlabel('Limites de Clase')
plt.ylabel('Frecuencia Acumulada')
plt.grid(True)
plt.legend()
plt.show()
```
